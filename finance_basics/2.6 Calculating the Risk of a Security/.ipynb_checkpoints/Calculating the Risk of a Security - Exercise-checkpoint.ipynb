## Calculating the Risk of a Security

*Suggested Answers follow (usually there are multiple ways to solve a problem in Python).*

Download the data for Microsoft (‘MSFT’) from Yahoo Finance for the period ‘2000-1-1’ until today. <br />
Assess the daily and the annual risk of ‘MSFT’. Repeat the exercise for Apple for the same timeframe. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
data_dir = '../dataset/'
tickers = ['MSFT','AAPL']
adj_prices = pd.DataFrame()
for t in tickers:
    adj_prices[t] = pd.read_csv(os.path.join(data_dir, t+'.csv'), index_col=0)['Adj Close']
adj_prices
```

```python
log_returns = np.log(adj_prices / adj_prices.shift(1))
log_returns
```

### MSFT

Daily risk:

```python
log_returns['MSFT'].std()
```

Annual risk:

```python
log_returns['MSFT'].std() * 250 ** 0.5
```

### Apple

Daily risk:

```python
log_returns['AAPL'].std()
```

Annual risk:

```python
log_returns['AAPL'].std() * 250 ** 0.5
```

Store the volatilities of the two stocks in an array called “vols”.

```python
vols = log_returns[['MSFT', 'AAPL']].std() * 250 **0.5
vols
```
